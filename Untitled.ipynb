```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://books.toscrape.com/catalogue/page-1.html'
response= requests.get(url)
```

```python
response=response.content
response
```

```python
soup=BeautifulSoup(response,'html.parser')
soup
```

```python
# finding the orderedlist ol which contains inforamtion about all the books and since  we want the first one we use find:
ol=soup.find('ol')
```

```python
# finding all the articles html tags with the class product_pod
articles=ol.find_all('article',class_='product_pod')
articles
```

```python
books=[]
```

```python
# we  see that the attribute value in the image tags have the full names of the book, so we want to grab that:
titles=[]
for article in articles:
    image=article.find('img')
    title=image.attrs['alt']
    print(title)
    titles.append(title)
    
```

```python
# Getting the star rating, which is stored in  class of the  first p tag inside the article:
stars=[]
for article in articles:
    star=article.find('p')
    star= star['class'][1]  # grabbing the name of the classes in the dictionary as there can be mutiple classes, but the class we want is in 
                            #  the second position, or index 1
    print(star)
    stars.append(star)
    
```

```python
# getting the price, which is stored in a p tag with a class:
prices=[]
for article in articles:
    # we want the text that is inside the p tags:
    price=article.find('p','price_color').text
    # converting to number and removing the  pound:
    price= float(price[1:])
    print(price)
    prices.append(price)
```

```python
# making a book list:
books=[]
lengthPrices=len(prices)
lengthStars=len(stars)
lengthTitles = len(titles)

if(lengthPrices>lengthStars and lengthPrices>lengthTitles):
    longestLength= lengthPrices
elif(lengthStars>lengthPrices and lengthStars>lengthTitles):
    longestLength=lengthStars
else:
    longestLength=lengthTitles

print(longestLength)
```

```python
books=[None]*longestLength
for i in range (longestLength):
    books[i]= ([titles[i],prices[i],stars[i]])
    # alternatively books.append(titles[i],prices[i],stars[i])

print(books)
```
